## Create a pandas dataframe

```python
import numpy as np
import pandas as pd


saved_cases = oscn.request.CaseList(types=types, years=years,counties=counties, stop=50)
columns = ['id','FiledStr','ClosedStr','County','Type', 'Judge', 'Docket','Source']
case_data = lambda c: [c.case_index, c.filed, c.closed, c.county, c.type, c.judge, c.docket, c.source]
%time all_cases = pd.DataFrame([case_data(c) for c in saved_cases], columns = columns)
all_cases.head()
```

```python
all_cases.count()
```

```python
df=all_cases

def expand_list(df, list_column, new_column): 
    lens_of_lists = df[list_column].apply(len)
    origin_rows = range(df.shape[0])
    destination_rows = np.repeat(origin_rows, lens_of_lists)
    non_list_cols = (
      [idx for idx, col in enumerate(df.columns)
       if col != list_column]
    )
    expanded_df = df.iloc[destination_rows, non_list_cols].copy()
    expanded_df[new_column] = (
      [item for items in df[list_column] for item in items]
      )
    expanded_df.reset_index(inplace=True, drop=True)
    return expanded_df

#usage
expanded_dataframe = expand_list(df,"Docket","Minutes")
expanded_dataframe
```

### Convert date strings to datevalues

```python
# convert the date from a string to a python DateTime type
from datetime import datetime

def pd_to_datetime(date_string):
    return pd.to_datetime(date_string) if date_string != '' else datetime(1970,1,1,0,0)

all_cases['Filed'] = all_cases['FiledStr'].apply(pd_to_datetime)
all_cases['Closed'] = all_cases['ClosedStr'].apply(pd_to_datetime)

# calculate the number of days to close each case
all_cases['Duration'] = (all_cases['Closed']-all_cases['Filed']).dt.days
```

## Remove cases without issues


```python

strType = lambda typeNotation: f'{type(typeNotation)}'

all_cases['IssueType'] = all_cases['Issues'].apply(strType)

all_cases = all_cases[all_cases.IssueType != "<class \'bool\'>" ]


all_cases.groupby(['Type']).count()


all_cases.count()
```

## Expand dispositions

```python
# This function takes a column of list values and expands it to a row per list element
def expand_list(df,list_values, new_column_name):   
    # create a map to all the existing non disposition data
    lens = list_values.apply(len)
    vals = range(df.shape[0])
    ilocations = np.repeat(vals, lens)
    # get the index of all non target columns
    cols = [i for i,c in enumerate(df.columns) if c != list_values.name ]
    # create a new expanded data frame and copy the old data
    new_df = df.iloc[ilocations, cols].copy()
    # add the new column with the flatten list
    new_df[new_column_name] = [item for items in list_values for item in items ]
    new_df.reset_index(inplace=True, drop=True)
    return new_df
    
```

```python
# expand the issues
df = expand_list(all_cases, all_cases['Issues'],'Issue')

issue_desc = lambda i:i['Issue'] if i else ""
df['IssueDescription'] = df['Issue'].apply(issue_desc)

# get the disposition list and expand it
df['Dispositions'] = list(item['dispositions'] for item in df['Issue'])
dispositions = expand_list(df,df['Dispositions'],'Disposed')
```

```python
# remove the date from dispostion text

import re

find_desc = re.compile(r'(.*)\,')
def disposed(disp): 
    return find_desc.match(disp['Disposed']).group(1) if ',' in disp['Disposed'] else disp['Disposed']

dispositions['DisposedDescription'] = dispositions['Disposed'].apply(disposed)

# Add party 

def get_party(disposed):
    return disposed['Defendant'] if disposed['Defendant'] else disposed['Respondent']

dispositions['Party'] = dispositions['Disposed'].apply(get_party)

dispositions.groupby(['Type','Judge'])['Judge'].count()
```

### Save documents

```python
from elasticsearch import Elasticsearch
from elasticsearch import helpers
es = Elasticsearch(http_compress=True)

keys = ['id','Filed', 'Closed', 'County', 'Type', 'Judge', 'Duration',
        'IssueDescription', 'DisposedDescription', 'Party', 'Source']

def filterKeys(doc):
    return {k: doc[k] for k in keys }

def doc_generator(df):
    df_iter = df.iterrows()
    for idx, document in df_iter:
        yield {
                "_index": 'dispositions',
                "_type": "_doc",
                "_id" : f"{document['id']}-{idx}",
                "_source": filterKeys(document),
            }
    raise StopIteration

helpers.bulk(es, doc_generator(dispositions))
```
